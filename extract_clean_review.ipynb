```python
for i in range(0,len(kakao_df)):
  colNum = 3
  while (colNum < kakao_df.shape[1]):
    try:
      int(kakao_df.iloc[i,colNum])
      kakao_df.iloc[i,2] = kakao_df.iloc[i,colNum-1]
      kakao_df.iloc[i,3] = kakao_df.iloc[i,colNum]
      kakao_df.iloc[i,4] = kakao_df.iloc[i,colNum+1]
      kakao_df.iloc[i,5] = kakao_df.iloc[i,colNum+2]
      break
    except ValueError:
      pass
    colNum+=2

for i in range(0,len(carrot_df)):
  colNum = 3
  while (colNum < carrot_df.shape[1]-3):
    try:
      int(carrot_df.iloc[i,colNum])
      carrot_df.iloc[i,2] = carrot_df.iloc[i,colNum-1] # review
      carrot_df.iloc[i,3] = carrot_df.iloc[i,colNum]   # rating
      carrot_df.iloc[i,4] = carrot_df.iloc[i,colNum+1] # false/true
      carrot_df.iloc[i,5] = carrot_df.iloc[i,colNum+3] # title
      break
    except ValueError:
      pass
    colNum+=2

kakao_df = kakao_df.iloc[:,0:7]
carrot_df = carrot_df.iloc[:,0:7]
```

```python
# Standardize the data by selecting shared columns (date, review, rating, isEdited, title) and placing data in the right columns

df_set = [kakao_df, netflix_df, coupangeats_df, youtube_df, elevenst_df, naver_df, coupang_df, instagram_df, baemin_df, carrot_df, tiktok_df]

for i in range(0,len(df_set)):
  try:
    df_set[i].drop(['userName'], axis=1, inplace=True)
  except KeyError:
    try:
      df_set[i].drop(['developerResponse'], axis=1, inplace=True)
    except KeyError:
      pass

kakao_dict = {'Unnamed: 1': 'date',
        'Unnamed: 2': 'review',
        'Unnamed: 3': 'rating',
        'Unnamed: 4': 'isEdited',
        'Unnamed: 5': 'title'}
df_set[0].rename(columns=kakao_dict,inplace=True)

for i in range(0,len(df_set)):
  try:
    df_set[i] = df_set[i][df_set[i].columns.drop(list(df_set[i].filter(regex='Unnamed')))]
  except KeyError:
    pass
  df_set[i] = df_set[i].iloc[:, 0:5]

df_set[2]
```

```python
final_df = df_set[0].append([df_set[x] for x in range(1,len(df_set))])
final_df.to_csv('/content/drive/MyDrive/cleanedReview.csv')
```
